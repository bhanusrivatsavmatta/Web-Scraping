```python
#create list of cities

city_list = ['Delhi','Mumbai','Chicago','Los_Angeles','Sydney','Seoul']




#automate
```

```python
import requests
from lxml import html
from bs4 import BeautifulSoup
import pandas as pd
import numpy as np
import re

def extract_data(link):
    r = requests.get(link)
    soup = BeautifulSoup(r.content, "html.parser")

    tables = soup.findAll("table", { "class" : "wikitable" })

    # show tables
    #for table in tables:
        #print ("###############")
        #print (table.text[:100])

    for tn in range(len(tables)):
        table=tables[tn]

        # preinit list of lists
        rows=table.findAll("tr")
        row_lengths=[len(r.findAll(['th','td'])) for r in rows]
        ncols=max(row_lengths)
        nrows=len(rows)
        data=[]
        for i in range(nrows):
            rowD=[]
            for j in range(ncols):
                rowD.append('')
            data.append(rowD)

    # process html
        for i in range(len(rows)):
            row=rows[i]
            rowD=[]
            cells = row.findAll(["td","th"])
            for j in range(len(cells)):
                cell=cells[j]

                #lots of cells span cols and rows so lets deal with that
                cspan=int(cell.get('colspan',1))
                rspan=int(cell.get('rowspan',1))
                for k in range(rspan):
                    for l in range(cspan):
                        data[i+k][j+l]+=cell.text

            data.append(rowD)
    
    frame = pd.DataFrame(data)
    frame = frame[1:7]
    for i in range(frame.shape[0]):
        for j in range(frame.shape[1]):
    
            frame.iloc[i,j]=frame.iloc[i,j].splitlines()[0]
    return(frame)
```

```python
link_list = []
for city in city_list:
    
    link_list.append("https://en.wikipedia.org/wiki/Climate_of_" + city)

link_list
```

```python
Delhi_Data = extract_data(link_list[0])
Chicago_Data = extract_data(link_list[2])
Sydney_Data = extract_data(link_list[4])
Seoul_Data = extract_data(link_list[5])
```

```python
#Data for Tableau

Seoul_Data = Seoul_Data.drop(Seoul_Data.index[3])
Chicago_Data = Chicago_Data.drop(Chicago_Data.index[3])
Delhi_Data = Delhi_Data.drop(Delhi_Data.index[5])
Sydney_Data = Sydney_Data.drop(Sydney_Data.index[5])


```

```python
Sydney_Data.rename(columns={'Sydney':'City'}, inplace=True)
Delhi_Data.rename(columns={'Delhi':'City'}, inplace=True)
Chicago_Data.rename(columns={'Chicago':'City'}, inplace=True)
Seoul_Data.rename(columns={'Seoul':'City'}, inplace=True)
```
